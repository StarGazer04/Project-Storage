**2.5 (1.5/5) Matrix Power**

Implement your own function first:

```python
# Problem 2.5 (a)
def mat_pow(A,p):
  # check input
  assert A.shape[0] == A.shape[1], "Matrix input needs to be square!"
  assert p > 0, "p should be a positive integer!"
  assert isinstance(p, int), "p should have the data type of integer!"

  ###### YOUR CODE STARTS HERE ######
  Ap = A
  for i in range(p-1):
    Ap = Ap @ A
  #    ....
  ###################################
  return Ap

# print (mat_pow(A, 2))
# np.linalg.matrix_power(A, 2)
```

Call your own function to solve the given questions:

```python
from numpy.matrixlib.defmatrix import mat
print("Problem 2.5(b)")
###### YOUR CODE STARTS HERE ######

solution_5_b_1 = mat_pow(A,6) @ mat_pow(A, 8) # put (A^6)(A^8)here

solution_5_b_2 = mat_pow(A, 6+8) # put A^(6+8) here

relative_dist_5_b = np.linalg.norm(solution_5_b_1 - solution_5_b_2) / (np.linalg.norm(solution_5_b_1))

###################################
print(solution_5_b_1)
print()
print(solution_5_b_2)
print()
print("relative distance between these two matrices is: ")
print(relative_dist_5_b)
```

**2.6 (1.5/5) Inverse and Transpose**

```python
print("Problem 2.6(a)")
###### YOUR CODE STARTS HERE ######

solution_6_a_1 = np.linalg.inv(A @ D) # put (AD)^(-1) here

solution_6_a_2 = np.linalg.inv(D) @ np.linalg.inv(A) # put (D^(-1))(A^(-1)) here

relative_dist_6_a = np.linalg.norm(solution_6_a_1 - solution_6_a_1) / np.linalg.norm(solution_6_a_1)

###################################
print(solution_6_a_1)
print()
print(solution_6_a_2)
print()
print("relative distance between these two matrices is: ")
print(relative_dist_6_a)
print()
```

```python
print("Problem 2.6(b)")
###### YOUR CODE STARTS HERE ######
solution_6_b_1 = (np.linalg.inv(A)).T # put (A^(-1))^T here

solution_6_b_2 = np.linalg.inv(A.T) # put (A^T)^(-1) here

relative_dist_6_b = np.linalg.norm(solution_6_b_1 - solution_6_b_1) / np.linalg.norm(solution_6_b_1)

###################################
print(solution_6_b_1)
print()
print(solution_6_b_2)
print()
print("relative distance between these two matrices is: ")
print(relative_dist_6_b)
print()
```
